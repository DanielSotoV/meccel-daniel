## Sistema séxtuple jerárquico de la estrella Castor

Castor (α Geminorum) no es una sola estrella, sino un **sistema estelar séxtuple** con una estructura jerárquica estable:

- **Castor A**: binaria espectroscópica (**Aa + Ab**).
- **Castor B**: binaria espectroscópica (**Ba + Bb**).
- **Castor C**: corresponde a **YY Geminorum**, también binaria cercana (**Ca + Cb**).

La jerarquía puede entenderse en niveles:

1. **Tres binarias internas** muy compactas: A, B y C.
2. **A y B** forman entre sí una binaria visual más amplia (par **AB**).
3. El par **AB** y el par **C (YY Gem)** orbitan un centro de masa común en una órbita aún más grande.

Este arreglo “binarias dentro de binarias” reduce perturbaciones dinámicas fuertes y permite que el sistema sea relativamente longevo. Por eso se clasifica como un ejemplo clásico de **sistema múltiple jerárquico**.

### Diagrama jerárquico (tipo *tree*)

```text
Castor (α Geminorum) [Sistema séxtuple]
├── Par AB
│   ├── Castor A
│   │   ├── Aa
│   │   └── Ab
│   └── Castor B
│       ├── Ba
│       └── Bb
└── Castor C (YY Gem)
    ├── Ca
    └── Cb
```

En forma compacta: **((Aa–Ab) + (Ba–Bb)) + (Ca–Cb)**.

## Sistema jerárquico de **30 Arietis**

**30 Arietis** es un sistema múltiple **jerárquico** (cuádruple) organizado en varios niveles orbitales:

- **30 Ari A**: binaria espectroscópica (**Aa + Ab**), muy compacta.  
- **30 Ari B**: estrella principal con:
    - un compañero estelar cercano (**30 Ari C**), y
    - un planeta gigante conocido (**30 Ari Bb**).

La jerarquía se entiende así:

1. Subsistemas internos compactos (Aa–Ab y B–C).  
2. La componente **A** y la componente **B** forman un par visual mucho más amplio.  

Esta separación de escalas (órbitas internas pequeñas y órbita externa grande) es la firma de un **sistema jerárquico**, porque ayuda a mantener la estabilidad dinámica a largo plazo.

### Diagrama jerárquico (tipo *tree*)

```text
30 Arietis [Sistema cuádruple jerárquico]
├── 30 Ari A
│   ├── Aa
│   └── Ab
└── 30 Ari B
    ├── 30 Ari C
    └── 30 Ari Bb (planeta gigante)
```

En forma compacta: **(Aa–Ab) + (B–C + Bb)**.

# Coordenadas de Jacobi

## Coordenadas de Jacobi para mecánica celeste

Las **coordenadas de Jacobi** son una forma de describir un sistema de $N$ cuerpos separando el movimiento global (centro de masa) del movimiento interno relativo.  
Son especialmente útiles en **mecánica celeste** porque simplifican la formulación del problema y su integración numérica.

### Idea general

Para masas $m_1, m_2, \dots, m_N$ con posiciones $\mathbf{r}_1, \mathbf{r}_2, \dots, \mathbf{r}_N$:

1. Se define primero el vector relativo del par interno:
    $$
    \boldsymbol{\rho}_1 = \mathbf{r}_2 - \mathbf{r}_1
    $$
2. Luego, cada nuevo vector une el siguiente cuerpo con el centro de masa de los anteriores:
    $$
    \boldsymbol{\rho}_2 = \mathbf{r}_3 - \frac{m_1\mathbf{r}_1 + m_2\mathbf{r}_2}{m_1 + m_2}
    $$
    $$
    \boldsymbol{\rho}_3 = \mathbf{r}_4 - \frac{m_1\mathbf{r}_1 + m_2\mathbf{r}_2 + m_3\mathbf{r}_3}{m_1 + m_2 + m_3}
    $$
    y así sucesivamente.
3. Además, se usa el **centro de masa total**:
    $$
    \mathbf{R}_{CM} = \frac{1}{M}\sum_{i=1}^{N} m_i \mathbf{r}_i,\qquad M=\sum_{i=1}^{N} m_i
    $$

### Ventajas

- Separan de manera natural el movimiento del centro de masa.
- Reducen acoplamientos innecesarios en la energía cinética.
- Son ideales para sistemas **jerárquicos** (binarias dentro de binarias).
- Facilitan la construcción de variables canónicas y métodos simplécticos.

### Interpretación física

En un sistema jerárquico, cada coordenada de Jacobi representa una “órbita relativa” en un nivel de la jerarquía:
- órbitas internas compactas,

Por eso son una herramienta estándar para estudiar estabilidad y evolución orbital en sistemas múltiples.

$$
M_i \equiv \sum_{k=1}^{i} m_k,\qquad M\equiv \sum_{k=1}^{N} m_k,\qquad 
\mathbf R_{CM}=\frac{1}{M}\sum_{k=1}^{N}m_k\mathbf r_k
$$

La coordenada de Jacobi $i$-ésima es
$$
\boldsymbol{\rho}_i
=\mathbf r_{i+1}-\frac{1}{M_i}\sum_{k=1}^{i}m_k\mathbf r_k.
$$

Si se quiere escribir explícitamente usando $\mathbf R_{CM}$:
$$
\sum_{k=1}^{i}m_k\mathbf r_k
= M\mathbf R_{CM}-\sum_{k=i+1}^{N}m_k\mathbf r_k,
$$
por tanto
$$
\boxed{
\boldsymbol{\rho}_i
=\mathbf r_{i+1}
-\frac{M\mathbf R_{CM}-\sum_{k=i+1}^{N}m_k\mathbf r_k}{M_i}
}.
$$

Caso particular $i=N-1$:
$$
\boldsymbol{\rho}_{N-1}
=\mathbf r_N-\frac{1}{M_{N-1}}\sum_{k=1}^{N-1}m_k\mathbf r_k
=\frac{M}{M_{N-1}}\big(\mathbf r_N-\mathbf R_{CM}\big).
$$

Indicamos la configuración del primer subsistema en coordenadas de Jacobi pero solo los vectores relativos

````python
m_Zul_Aa = 1
m_Zul_Ab = 0.01

r_Zul_A = np.array([0.1,0,0])
v_Zul_A = np.array([0.1,0,0])
````

````python
m_Zul_Aab = m_Zul_Aa + m_Zul_Ab
m_Zul_B = 1.0

r_Zul_AB = np.array([1.0,0,0])
v_Zul_AB = np.array([0,1.0,0])
````

Armar el sistema:

Comenzamos definiendo el centro de masa de todo el sistema

````python
RCM_Zul_AB = np.array([0,0,0])
VCM_Zul_AB = np.array([0,0,0])
````

Descendemos por el árbol del sistema jerárquico

````python
# Este es el equivalente a la particula 1

RCM_Zul_A = RCM_Zul_AB + m_Zul_B/(m_Zul_Aab + m_Zul_B)*r_Zul_AB
VCM_Zul_A = VCM_Zul_AB + m_Zul_B/(m_Zul_Aab + m_Zul_B)*v_Zul_AB


# Este es el equivalente a la particula 2

r_Zul_B = RCM_Zul_AB - m_Zul_Aab/(m_Zul_Aab + m_Zul_B)*r_Zul_AB
v_Zul_B = VCM_Zul_AB - m_Zul_Aab/(m_Zul_Aab + m_Zul_B)*v_Zul_AB
````

En el siguiente nivel del árbol:

````python
# Particula 1

r_Zul_Aa = RCM_Zul_A + m_Zul_Ab/m_Zul_Aab*r_Zul_A
v_Zul_Aa = VCM_Zul_A + m_Zul_Ab/m_Zul_Aab*v_Zul_A

# Particula 2

r_Zul_Ab = RCM_Zul_A - m_Zul_Aa/m_Zul_Aab*r_Zul_A
v_Zul_Ab = VCM_Zul_A - m_Zul_Aa/m_Zul_Aab*v_Zul_A
````
