# Fuerzas de van der Waals

Las fuerzas de van der Waals son fuerzas intermoleculares débiles que surgen de interacciones entre átomos y moléculas. Se pueden clasificar en tres tipos principales:

## 1. Dipolo-Dipolo (Keesom)

Para moléculas polares:

$$U = -\frac{2\mu_1^2 \mu_2^2}{3k_B T r^6}$$

donde $\mu_1$ y $\mu_2$ son los momentos dipolares, $k_B$ es la constante de Boltzmann, $T$ es la temperatura y $r$ es la distancia.

## 2. Dipolo Inducido (Debye)

Cuando una molécula polar induce un dipolo en otra no polar:

$$U = -\frac{\mu_1^2 \alpha_2}{r^6}$$

donde $\alpha_2$ es la polarizabilidad de la segunda molécula.

## 3. Dispersión (London)

Para moléculas no polares:

$$U = -\frac{C}{r^6}$$

donde $C = \frac{3\alpha_1 \alpha_2 I_1 I_2}{2(I_1 + I_2)}$, siendo $\alpha_i$ las polarizabilidades e $I_i$ los potenciales de ionización.

## Forma General

La energía potencial de van der Waals frecuentemente se expresa como:

$$U(r) = \frac{A}{r^{12}} - \frac{B}{r^6}$$

o mediante el potencial de Lennard-Jones:

$$U(r) = 4\epsilon \left[\left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^6\right]$$

donde $\epsilon$ es la profundidad del pozo de potencial y $\sigma$ es la distancia de equilibrio.
